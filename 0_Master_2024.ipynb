## 1. Imports

```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print('Libraries loaded successfully.')
```

## 2. Set Base Directory

⚠️ **Change `BASE_DIR` to your own path before running.**

```python
# ── CHANGE THIS ─────────────────────────────────────────────────────────────
BASE_DIR = "C:/Users/Utente/OneDrive/Immagini/Documenti/Economic_Risk_work"   # e.g. "C:/Users/YourName/HHEc/" on Windows
# ────────────────────────────────────────────────────────────────────────────

# Global path variables (equivalent to Stata globals)
PATHS = {
    "DataIn":   os.path.join(BASE_DIR, "DataIn"),
    "DataOut":  os.path.join(BASE_DIR, "DataOut"),
    "programs": os.path.join(BASE_DIR, "programs"),
    "graphs":   os.path.join(BASE_DIR, "graphs"),
    "logfiles": os.path.join(BASE_DIR, "logfiles"),
}

print("Paths configured:")
for k, v in PATHS.items():
    print(f"  {k:10s} → {v}")
```

## 3. Create Folder Structure
Equivalent to Stata's `cap mkdir` commands.

```python
for name, path in PATHS.items():
    os.makedirs(path, exist_ok=True)   # equivalent to 'cap mkdir' (no error if exists)
    print(f"  ✓  {path}")

print("\nFolder structure ready.")
```

## 4. Verify Input Data
Check that the raw `.dta` files from Banca d'Italia are present in `DataIn`.

```python
REQUIRED_FILES = [
    "comp.dta", "cons.dta", "rfam.dta", "ricf.dta",
    "peso.dta", "defl.dta", "ldip.dta"
]

print(f"Checking DataIn: {PATHS['DataIn']}\n")
all_ok = True
for f in REQUIRED_FILES:
    fpath = os.path.join(PATHS['DataIn'], f)
    status = "✓  found" if os.path.exists(fpath) else "✗  MISSING"
    if "MISSING" in status:
        all_ok = False
    print(f"  {status}  →  {f}")

print()
if all_ok:
    print("All required files found. You can proceed to notebook 1_merge.")
else:
    print("⚠️  Some files are missing. Download the SHIW storico_stata.zip from Banca d'Italia and extract it into DataIn.")
```

## 5. Export PATHS for Other Notebooks
Save the paths to a small JSON config file so notebooks 1 and 2 can import them automatically.

```python
import json

config_path = os.path.join(BASE_DIR, "config.json")
with open(config_path, "w") as f:
    json.dump(PATHS, f, indent=2)

print(f"Config saved to: {config_path}")
print("\nNotebook 0 — Master Setup complete ✓")
```
